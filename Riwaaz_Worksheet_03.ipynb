# Problem 5 -

```python
#Calculating total fare

total_fare = titanic["Fare"].sum()
total_fare
```

## total fare for each class

```python
fare_class_1 = titanic[titanic["Pclass"] == 1]["Fare"].sum()
fare_class_2 = titanic[titanic["Pclass"] == 2]["Fare"].sum()
fare_class_3 = titanic[titanic["Pclass"] == 3]["Fare"].sum()
```

## Combine into a list

```python
fare_totals = [fare_class_1, fare_class_2, fare_class_3]
fare_totals
```

## Calculate proportions

```python
proportions = [(fare / total_fare) * 100 for fare in fare_totals]
proportions
```

## Display results

```python
print("Percentage of total fare revenue by class:")
print(f"Class 1: {proportions[0]:.2f}%")
print(f"Class 2: {proportions[1]:.2f}%")
print(f"Class 3: {proportions[2]:.2f}%")
```

# Problem 6 - Percentage of passengers belonging to each age group

```python
titanic["Age"] = titanic["Age"].fillna(titanic["Age"].median())
```

## Create the age group column

```python
def classify_age(age):
    if age < 18:
        return "child"
    elif 18 <= age < 65:
        return "adult"
    else:
        return "senior"

titanic["age_group"] = titanic["Age"].apply(classify_age)
```

## Total number of passengers

```python
total_passengers = len(titanic)
total_passengers
```

## Count number in each age group


```python
age_group_counts = titanic["age_group"].value_counts()
age_group_counts
```

## Calculate percentage

```python
age_group_percent = (age_group_counts / total_passengers) * 100
age_group_percent
```

## Display output - clean

```python
print("Percentage of passengers in each age group:")
for group, percent in age_group_percent.items():
    print(f"{group.capitalize()}: {percent:.2f}%")
```
